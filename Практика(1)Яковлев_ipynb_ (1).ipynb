## **4. Визуализация распределения переменных**

### **Описание:**
Создайте визуализации для понимания распределения числовых и категориальных переменных.

### **Шаги:**


1. **Гистограмма для числовых переменных:**

```python
num_cols = df.select_dtypes(include='int64').columns # если есть int чиловая переменная либо + ['col1', 'col2'], либо такие колонки преобразовать во flaot
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()

for ax, col in zip(axes, num_cols):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(f'Распределение {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Частота')


plt.tight_layout()
plt.show()
```

Нормальное распределение:
Hours_Studied, Sleep_Hours, Physical_Activity, Tutoring_Sessions, Exam_Scores
Ненормальное распрделение:
Attendance, Previous_Scores

Концентрация большинства значений:
Hours_Studied: два основных кластера - около 18 часов и около 22 часов
Attendance: концентрация между 65-75% и 90-95% (два выраженных пика)
Sleep_Hours: пик вокруг значения 7 часов
Previous_Scores: нормальное распределение вокруг 75-80 баллов
Tutoring_Sessions: резкий пик на 0-1 сессиях, быстрое убывание
Physical_Activity: концентрация в диапазоне 3-4 балла
Exam_Score: нормальное распределение вокруг 67-70 баллов

Предсказание:
Exam_Score → School_Type: Четкое разделение - студенты частных школ в среднем получают на 2-3 балла выше, распределения почти не пересекаются в области высоких оценок

Hours_Studied → Motivation_Level: Идеальный предиктор - Low motivation: 10-18ч, Medium: 17-23ч, High: 22-28ч, практически без перекрытия между крайними группами

Attendance → Learning_Disabilities: Сильная дискриминация - студенты с нарушениями обучения почти никогда не имеют посещаемость выше 90%, концентрируются в диапазоне 60-80%

**Боксплот для числовой переменной:**

```python
num_cols = df.select_dtypes(include='int64').columns # если есть int чиловая переменная либо + ['col1', 'col2'], либо такие колонки преобразовать во flaot
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()
for ax, col in zip(axes, num_cols):
    sns.boxplot(df[col], ax=ax)
    ax.set_title(f'Распределение {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Частота')

plt.tight_layout()
plt.show()
```

Hours_Studied: Много выбросов сверху и снизу - студенты заучки и лентяи

Tutoring_Sessions: Выбросы сверху - интенсивная подготовка

Exam_Score/Previous_Scores: Немного выбросов с обеих сторон - естественный разброс способностей


Hours_Studied: Low motivation - 16ч, High - 25ч (+9ч)
Previous_Scores: High School parents - 68, Postgraduate - 87 (+19 баллов)
Attendance: No disabilities - 85%, Yes - 70% (-15%)
Exam_Score: Public school - 67, Private - 70 (+3 балла)
Sleep_Hours: Near home - 7.2ч, Far - 6.7ч (-0.5ч)

**Постройте корреляционную матрицу для всех числовых переменных и визуализируйте её с помощью тепловой карты:**

```python
# Корреляционная матрица
numeric_columns = df.select_dtypes(include=['int64'])
correlation_matrix = numeric_columns.corr()

# Визуализация
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```

Исходя из корреляционной матрицы, у нас ни одной сильной корреляции и почти все находится в отрицательном.
Лишь зависимости между Exam_Score, Hours, Studied, Attendance хоть какие-то(около 0.5)
